```python
url ='https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv'
crime = pd.read_csv(url)
crime.head()
```

```python
crime.head()
```

```python
crime.index
```

```python
crime.loc[[1960,1961]]
```

```python
#crime.iloc[[3,4,8],[crime.loc(axis =1)['Murder','Robbery']]]
crime.loc[crime.index[[3,4,8]],['Murder','Robbery']]
```

```python
# select only the rows where the number of murder is greater than 24,000
crime[crime.Murder > 24000]
#crime[crime['Murder'] > 24000]
```

```python
# select rows where murder is between 20000 and 24000 (inclus
crime[crime['Murder'].between(20000,24000)].head()
```

```python
#calculate the mean murder for each different year in df
crime.groupby('Year')['Murder'].mean()
```

```python
# sort df first by the values in the Murder in decending order, then by the 
#value in the Violent column in ascending order
crime.sort_values(['Murder','Violent'],ascending=[False,True]).head()
```
